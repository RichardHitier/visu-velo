```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

months_idx = pd.date_range(start='2024-01-15', end='2024-10-15', freq=pd.offsets.MonthBegin())
monday_idx = pd.date_range(start='2024-01-15', end='2024-10-15', freq=pd.offsets.Week(weekday=0))
_df = pd.DataFrame(index=monday_idx, data = len(monday_idx)*[0])

fig, ax = plt.subplots(2, figsize=(20,8), sharex=True)

ax1=ax[1]

label_fontsize=15
ax1.tick_params(axis='x', labelsize=label_fontsize-5, rotation=30)
date_format = '%d %b'
ax1.xaxis.set_major_formatter(mdates.DateFormatter(date_format))
ax1.set_xticks(monday_idx)
ax1.set_xticklabels (ax1.get_xticklabels(), ha="right")
ax1.set_xlabel("Lundis", fontsize=label_fontsize, labelpad=9.0, loc="left")

# sec2 = ax1.secondary_xaxis(location=-0.18)
# sec2.xaxis.set_major_formatter(mdates.DateFormatter('%m'))
# sec2.set_xticks(months_idx)
# o
sec2 = ax1.secondary_xaxis(location=-0.23)
sec2.xaxis.set_major_locator(mdates.MonthLocator(bymonthday=1))

# note the extra spaces in the label to align the month label inside the month.
# Note that this could have been done by changing ``bymonthday`` above as well:
sec2.xaxis.set_major_formatter(mdates.DateFormatter('  %b'))
sec2.tick_params('x', length=0, labelsize=12)
sec2.spines['bottom'].set_linewidth(0)
# sec2.set_xticklabels (sec2.get_xticklabels(), ha="center")
# sec2.tick_params('x', length=40, width=1.5)

for monday in monday_idx:
    ax1.axvline(monday,color='black', lw=0.5, alpha=0.5, linestyle='-')

ax1.plot(_df.index, _df)
plt.show()
```
